```python
#initial cell
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib

#make data
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
z = np.arange(1,101)
```

```python
#draw a scatter plot of y vs x in a bigger canvas
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y)
```

```python
#plot the marker size based on z values
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y,s=z)
```

```python
#now add a legend for marker sizes
```

```python
fig,ax = plt.subplots(figsize = (10,10))
p1 = ax.scatter(x,y,s=z)
handles,labels = p1.legend_elements(prop = "sizes",alpha = 0.5)
ax.legend(handles,labels)
```
