```python
#initial cell
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib

#make data
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
z = np.arange(1,101)
```

```python
#draw a scatter plot of y vs x in a figure sized 10*10
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y)
```

```python
#colour the markers based on z values
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y,c=z)
```

```python
#now add a legend for marker colors
```

```python
fig,ax = plt.subplots(figsize = (10,10))
p1 = ax.scatter(x,y,c=z)
handles,labels = p1.legend_elements(prop = "colors")
ax.legend(handles,labels)
```
