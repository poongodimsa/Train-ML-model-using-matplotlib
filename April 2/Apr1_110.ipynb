```python
#initial cell
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib
import random

#make data
x= np.random.normal(1,.75,10)
bins=np.arange(11)
```

```python
#draw a step plot and stair plot of x with bins 
```

```python
fig,ax = plt.subplots()
ax.stairs(x,bins)
ax.step(bins[:-1], x)
```

```python
#add a legend
```

```python
fig,ax = plt.subplots()
ax.stairs(x,bins, label = 'Stair plot')
ax.step(bins[:-1], x,  label = 'Step plot')
ax.legend()
```

```python
#please move the stair plot little upwards
```

```python
fig,ax = plt.subplots()
ax.stairs(x-1,bins, label = 'Stair plot')
ax.step(bins[:-1], x, label = 'Step plot')
ax.legend()
```
