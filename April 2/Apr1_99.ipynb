```python
#initial cell
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib

#make data
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
price = np.random.randint(1,100,100)
rank = np.random.randint(1,100,100)
```

```python
#draw a scatter plot of y vs x in a large sized figure
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y)
```

```python
#use the color map 'summer' and colour the markers by price
```

```python
fig,ax = plt.subplots(figsize = (10,10))
ax.scatter(x,y, c = price, cmap = 'summer')
```

```python
#now set the size of marker based on rank with a legend
```

```python
fig,ax = plt.subplots(figsize = (10,10))
p1 = ax.scatter(x,y, c = price, cmap = 'summer', s = rank)
handles,labels = p1.legend_elements(prop = "sizes")
ax.legend(handles,labels)
```
