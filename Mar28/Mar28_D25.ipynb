```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np
import matplotlib.patches as patches

# make data
price  = [1200,3400,2000,5890,4789]
model = ['I am Volkswagon','I am Mercedes Benz','I am Maruti Suzuki','I am TATA','I am a BMW']

```

```python
# Make a bar chart of price and model
```

```python
fig, ax = plt.subplots()
plt.bar(model,price)
```

```python
fig, ax = plt.subplots()
plt.bar(model,price)
plt.xticks(rotation = 90)
```
