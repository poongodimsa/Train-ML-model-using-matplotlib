```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np
import matplotlib.patches as patches

# make data
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - ( np.random.normal(0, 0.1, 1000)) * 3
```

```python
# make scatter plot of purchaseAmount vs pageSpeeds
```

```python
fig, ax = plt.subplots()
plt.scatter(pageSpeeds, purchaseAmount)
```

```python
#decrease the marker size
```

```python
fig, ax = plt.subplots()
plt.scatter(pageSpeeds, purchaseAmount, s=3)
```

```python
#Add a square to highlight the centre area 
```

```python
fig, ax = plt.subplots()
plt.scatter(pageSpeeds, purchaseAmount, s = 3)
ax.add_patch(patches.Rectangle(
(2.5,99.75),
1,.5, alpha=0.5, facecolor = 'yellow'))
```
