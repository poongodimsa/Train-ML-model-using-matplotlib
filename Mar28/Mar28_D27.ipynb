```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np
import matplotlib.patches as patches

# make data
x = np.arange(1,50,2)
y =  np.cos(x)
```

```python
# plot x vs y with labels
```

```python
fig, ax = plt.subplots()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#add a horizontal line at 0
```

```python
fig, ax = plt.subplots()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,0)
```

```python
#now make the curve yellow 
```

```python
fig, ax = plt.subplots()
plt.plot(x,y, color = 'yellow')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,0)
```
