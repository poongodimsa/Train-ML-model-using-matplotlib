```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np
import matplotlib.patches as patches

# make data
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - ( pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
```

```python
# make scatter plot of purchaseAmount vs pageSpeeds with marker size = 3
```

```python
fig, ax = plt.subplots()
plt.scatter(pageSpeeds, purchaseAmount, s = 2)
```
