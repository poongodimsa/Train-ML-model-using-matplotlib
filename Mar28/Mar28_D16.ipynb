```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
x = np.arange(.1,1,.01)
s = 1 + np.sin(2 * np.pi * x)
c = 1 + np.cos(2 * np.pi * x)
xtck = np.arange(0.1,1.1,.1)
```

```python
#plot s as funtion of x
```

```python
fig, ax = plt.subplots()
plt.plot(x,s)
```

```python
#Increment the ticks in steps of 0.1
```

```python
fig, ax = plt.subplots()
plt.plot(x,s)
ax.set_xticks(xtck)
```

```python
# add c to the plot
```

```python
fig, ax = plt.subplots()
plt.plot(x,s)
plt.plot(x,c)
ax.set_xticks(xtck)
```
