```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np
import matplotlib.patches as patches

# make data
x = np.arange(1,50,2)
z1 = x + np.sin(x)
z2 = 2*x + np.cos(x)
z3 = 2*x + np.sin(x)
```

```python
# create a stacked area chart of z1, z2, and z3
```

```python
fig, ax = plt.subplots()
plt.stackplot(x,z1,z2,z3)
```

```python
#now add a legend at upper left corner
```

```python
fig, ax = plt.subplots()
plt.stackplot(x,z1,z2,z3, labels = ['Z1','Z2','Z3'])
plt.legend(loc = 'upper left')
```

```python
#please change the colors of the stacks
```

```python
fig, ax = plt.subplots()
plt.stackplot(x,z1,z2,z3, labels = ['Z1','Z2','Z3'], colors = ["#9b59b6", "#e74c3c", "#34495e"])
plt.legend(loc = 'upper left')
```
