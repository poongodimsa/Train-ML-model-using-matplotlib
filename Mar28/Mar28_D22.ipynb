```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
t = np.arange(1,25,1) 
t2 = t**2
t4 = t**4
```

```python
# Plot t2 and t4 as a function of t
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(t2,t)
ax2.scatter(t4,t)
```

```python
#Add title to both the plots
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(t2,t)
ax2.scatter(t4,t)
ax1.set_title('Time squared')
ax2.set_title('Time to the fourth power')
```

```python
#Add a super title 'Time Functions' and make fontsize to 16
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.scatter(t2,t)
ax2.scatter(t4,t)
ax1.set_title('Time squared')
ax2.set_title('Time to the fourth power')
fig.suptitle('Time Functions', fontsize = 16)
```
