```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
pagespeeds = np.random.normal(3.0, 1.0, 10000)
purchaseamount = np.random.normal(50.0, 10.0, 10000) * pagespeeds
```

```python
#I need a sctter plot of  purchaseamount as a function of pagespeeds
```

```python
fig, ax = plt.subplots()
plt.scatter(pagespeeds, purchaseamount)
```

```python
fig, ax = plt.subplots()
plt.scatter(pagespeeds, purchaseamount)
ax.set_xlabel('Page Speed')
ax.set_ylabel('Purchase Amount')
```
