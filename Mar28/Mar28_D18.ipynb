```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
x = np.random.binomial(100,.5,100)
y = np.random.normal(100,size = 100)
```

```python
# Draw a violin plot for x
```

```python
fig, ax = plt.subplots()
plt.violinplot(x)
```

```python
#Draw a violin plot for y in a separate axes
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.violinplot(x)
ax2.violinplot(y)
```

```python
#Name x and y plot 'Binomial' and 'Normal' respectively
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.violinplot(x)
ax2.violinplot(y)
ax1.set_title('Binomial')
ax2.set_title('Normal')
```
