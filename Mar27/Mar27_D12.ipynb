```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
x = 0.15 + np.arange(50) 
y = np.random.uniform(1,15, len(x))
z = np.random.normal(1,15, len(x))
```

```python
#plot y and z as function of x
```

```python
fig, ax = plt.subplots()
plt.plot(x, y)
plt.plot(x, z)
```

```python
#Add legend
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, label = 'y')
plt.plot(x, z , label = 'z')
plt.legend(loc='upper left')
```

```python
#position the legend right
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, label = 'y')
plt.plot(x, z , label = 'z')
plt.legend(loc='upper right')
```
