```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
x = 0.15 + np.arange(50) 
y = np.random.uniform(1,15, len(x))
sizes = np.random.uniform(10,200,len(x))
```

```python
#Make scatter plot using x and y
```

```python
fig, ax = plt.subplots()
plt.scatter(x,y)
```

```python
#Add sizes to the plot
```

```python
fig, ax = plt.subplots()
plt.scatter(x,y,s=sizes)
```

```python
#Change the colour to green
```

```python
fig, ax = plt.subplots()
plt.scatter(x, y, s=sizes, color = 'g' )
```
