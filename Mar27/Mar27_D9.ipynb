```python
#Initial cell
%matplotlib inline
import pylab as plt
import numpy as np

# make data
x = 0.15 + np.arange(50) 
y = np.random.uniform(1,15, len(x))
```

```python
#plot a stem chart of x vs y
```

```python
fig, ax = plt.subplots()
plt.stem(x,y)
```

```python
#show a histogram of y adjacent to it
```

```python
fig, ax = plt.subplots(1,2)
ax[0].stem(x,y)
ax[0].set_title('Stem Chart')
ax[1].hist(y)
```

```python
#make the histogram magenta
```

```python
fig, ax = plt.subplots(1,2)
ax[0].stem(x,y)
ax[0].set_title('Stem Chart')
ax[1].hist(y, color = 'm')
```
